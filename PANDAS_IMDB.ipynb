```python
import pandas as pd
```

```python
dfa = pd.read_csv("celeb_data.csv")
pd.set_option('display.max_rows', 100)  #
pd.set_option('display.max_columns', 100)
```

```python
dfa.info()
```


*   Há espaços vazios nas colunas "height", "awards" e "birth place".





INICIANDO O TRATAMENTO DOS DADOS

```python

dfa['Date_of_birth'] = pd.to_datetime(dfa['Date_of_birth'])
dfa['Birth_year'] = dfa['Date_of_birth'].dt.year
```

*   Transformando a coluna 'Date_of_birth' para um tipo de dado "datetime AAAA-MM-DD" para poder tirar o ano de nascimento de cada ator.


```python

dfa.loc[24, "Birth_place"] = "Londres, Reino Unido" #Kingsley Ben-Adir
dfa.loc[75, "Birth_place"] = "Dallas, Texas, USA" #Miss Benny

dfa.loc[75, "height"] = '''5' 8"(1.73 m)'''#Miss Benny
dfa.loc[77, "height"] = '''5' 11"(1.80 m)'''#Christopher McQuarrie
dfa.loc[79, "height"] = '''5' 6"(1.73 m)'''#Ayo Edebiri

#dfa.loc[dfa["Awards"].isnull(), "Awards"] = 0

```




*   Atualizando alguns valores "NaN" do dataset.





```python

dfa['height_meters'] = dfa['height'].str.extract(r'(\d+\.\d+) m')
dfa['height_meters'] = dfa['height_meters'].astype(float)


#5′ 8″ (1.73 m)
```



*   Tirando o valor de metros da coluna"height", transformando-os em um tipo float e jogando em uma nova tabela "height_meters".



```python
dfa.loc[dfa["height"].isnull(), "height"] = 0
dfa.loc[dfa["height_meters"].isnull(), "height_meters"] = 0
```




*   Zerando alguns valores "NaN" do dataset.



```python
dfa['Awards'] = dfa['Awards'].str.replace('(\d+ m)', '0')
```



*   Zerando as linhas com dados errados(63 m) que estão presentes na coluna "Awards" usando o replace para encontrar os valores que correspondem ao padrão (\d+ m) e substitui-los por zero.


```python
dfa.loc[dfa["Awards"].isnull(), "Awards"] = 0
```



*   Zerando todos os "NaN" da coluna "Awards".


```python
dfa['Awards'] = dfa['Awards'].astype(str)

dfa['Wins'] = dfa['Awards'].str.extract('(\d+) win')

dfa['Wins'] = dfa['Wins'].fillna(0)

dfa['Wins'] = dfa['Wins'].astype(int)
```



*   Extraindo o número de vitórias e adicionando uma coluna "Wins".

*   Preenchendo com 0 os valores nulos.

*   Convertendo para o tipo int.



```python

dfa['Awards'] = dfa['Awards'].astype(str)

dfa['Nominations'] = dfa['Awards'].str.extract('(\d+) nomination')

dfa['Nominations'] = dfa['Nominations'].fillna(0)

dfa['Nominations'] = dfa['Nominations'].astype(int)
```



* Extraindo o número de nomeações e adicionando em uma coluna "Nominations".

* Preenchendo com 0 os valores nulos.

* Convertendo para um tipo int.



```python
dfa['Country_Birth'] = dfa['Birth_place'].str.split(',').str[-1].str.strip()
```



*   Retirando apenas o país de nascimento e preenchendo em uma nova tabela "Country_Birth".
